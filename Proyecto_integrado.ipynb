```python
# prompt: # Importar librerías necesarias
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
# # Cargar datos desde GitHub
# # Reemplazar con tu URL de GitHub raw
# url = 'URL_DE_TU_ARCHIVO_CSV_EN_GITHUB'
# df = pd.read_csv(url)
# # Crear gráfico de líneas para todas las categorías
# plt.figure(figsize=(15, 8))
# for columna in df.columns[1:]:  # Excluimos la columna 'Mes'
#     plt.plot(df['Mes'], df[columna], marker='o', label=columna)
# plt.title('Ventas de Productos Tecnológicos 2023')
# plt.xlabel('Mes')
# plt.ylabel('Unidades Vendidas')
# plt.xticks(rotation=45)
# plt.legend()
# plt.grid(True)
# plt.tight_layout()
# plt.show()
# # Espacio para análisis adicional del alumno
# # [AQUÍ EL ALUMNO PUEDE AGREGAR MÁS CÓDIGO]
# """
# Ideas para análisis adicional:
# 1. Calcular estadísticas descriptivas
# 2. Crear gráficos de barras para comparar categorías
# 3. Analizar tendencias estacionales
# 4. Calcular correlaciones entre productos
# 5. Crear visualizaciones personalizadas
# """

# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos desde GitHub
# Reemplazar con tu URL de GitHub raw
url = 'https://raw.githubusercontent.com/smyegros3/-proyecto_integrado_1/refs/heads/main/ventas_tecnologia.csv%20-%20Hoja%201.csv'  # Reemplaza con tu URL
df = pd.read_csv(url)

# Crear gráfico de líneas para todas las categorías
plt.figure(figsize=(15, 8))
for columna in df.columns[1:]:  # Excluimos la columna 'Mes'
    plt.plot(df['Mes'], df[columna], marker='o', label=columna)
plt.title('Ventas de Productos Tecnológicos 2023')
plt.xlabel('Mes')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Análisis adicional: Estadísticas descriptivas
print("Estadísticas descriptivas:")
print(df.describe())

# Análisis adicional: Gráfico de barras para comparar categorías (por ejemplo, promedio mensual)
promedio_mensual = df.iloc[:, 1:].mean()
plt.figure(figsize=(10, 6))
plt.bar(promedio_mensual.index, promedio_mensual.values)
plt.title('Promedio de Unidades Vendidas por Producto (2023)')
plt.xlabel('Producto')
plt.ylabel('Promedio de Unidades Vendidas')
plt.xticks(rotation=45)
plt.show()

# Análisis adicional: Correlación entre productos
correlation_matrix = df.iloc[:, 1:].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre Productos')
plt.show()

# Análisis adicional: Tendencias estacionales (ejemplo: promedio móvil)
df['Promedio Movil Laptops'] = df['Laptops'].rolling(window=3).mean()
plt.figure(figsize=(12, 6))
plt.plot(df['Mes'], df['Laptops'], label='Ventas Laptops')
plt.plot(df['Mes'], df['Promedio Movil Laptops'], label='Promedio Movil Laptops', linestyle='--')
plt.title('Ventas de Laptops con Promedio Movil')
plt.xlabel('Mes')
plt.ylabel('Unidades Vendidas')
plt.legend()
plt.show()
```
